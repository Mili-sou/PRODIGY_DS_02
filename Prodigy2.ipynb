# Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Load the dataset 

```python
titanic_data = pd.read_csv('train.csv')
titanic_data.head()
```

# Overview of the data

```python
# Get an overview of the dataset
titanic_data.info()
```

```python
# Generate summary statistics of the dataset
titanic_data.describe()
```

```python
# Check the shape of the dataset (number of rows, number of columns)
titanic_data.shape
```

# Check for NULL values

```python
# Check for missing values in each column
titanic_data.isnull().sum()
```

# Handling Missing/NULL values

```python
#Fill missing values in the 'Age' column with the median:
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)

# Drop columns with excessive missing data (e.g., Cabin)
titanic_data.drop('Cabin', axis=1, inplace=True)

# Drop rows with missing values in specific columns (e.g., Embarked)
titanic_data.dropna(subset=['Embarked'], inplace=True)
```

# Visualize the dataset and perform EDA

```python
# Explore the distribution of the variable
sns.countplot(x='Fare', data=titanic_data)
plt.show()
```

```python
#relationship between variables 
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.show()
```

```python
sns.countplot(x='Survived',hue='Pclass',data=titanic_data)
```

```python
sns.catplot(x='Survived', col='Sex', kind='count', data=titanic_data)
plt.show()
```

```python
sns.violinplot(x='Sex',y='Age',hue='Survived',data=titanic_data,split=True)
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age',data=titanic_data,palette='summer')
```

```python
# Analyze the age distribution of passengers
sns.histplot(titanic_data['Age'], bins=20,color="lightblue")
plt.xlabel('Age')
plt.ylabel('Count')
plt.title("Count of people based on age")
plt.show()
```

```python
# Plotting pairplot
sns.pairplot(titanic_data,hue='Survived',palette='BuPu')
plt.show()
```
